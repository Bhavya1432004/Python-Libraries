```python
# wac to change grade to 'dist' if per>=70, to 'first class' if per>=60, to 'second class' if per>=50 , otherwise 'fail' 
def change(n) :
    if n>=70 :
        return "DIST"
    elif n>=60 :
        return "FIRST"
    elif n>=50 :
        return "Second"
    else :
        return 'fail'
        
```

```python
df['grade'] = df['per'].apply(change)
df
```
