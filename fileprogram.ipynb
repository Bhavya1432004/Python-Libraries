```python
import xml.etree.ElementTree as et
```

```python
file = et.parse("mycollege.xml")
```

```python
root  = file.getroot()
root
```

```python
for record in root :
    vid = record.find("id").text
    vna = record.find("name").text
    vpe = record.find("per").text
    print(vid, " , ",vna," , ",vpe)
```

```python
rows = []
```

```python
for record in root :
    vid = record.find("id").text
    vna = record.find("name").text
    vpe = record.find("per").text
    rows.append({'id' : vid, 'name':vna, 'per': vpe})
```

```python
df = pd.DataFrame(rows)
df
```
